```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# importing the Salary Data Set
df = pd.read_csv(r'C:\Users\Loryne\Documents\LUXDEVHQ-LoryneJoy\DATA_SCIENCE\MACHINE-LEARNING\ML_LINEAR_REGRESSION\Salary_dataset.csv')
df.head()
```

```python
#to delete the unnamed column
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
#To fit the model
X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)# trains the machine

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

```python
# To test model to make predictions:
df['Predicted'] = model.predict(X)

plt.scatter(df['YearsExperience'], df['Salary'], label='Actual', color='blue')
plt.plot(df['YearsExperience'], df['Predicted'], label='Regression Line', color='red')
plt.title('Simple Linear Regression')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mse = mean_squared_error(df['Salary'], df['Predicted'])
mae = mean_absolute_error(df['Salary'], df['Predicted'])
rmse = np.sqrt(mse)
r2 = r2_score(df['Salary'], df['Predicted'])

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```
